```python
df['Base Price'] = pd.to_numeric(df['Base Price'], errors='coerce')
df['Winning Bid'] = pd.to_numeric(df['Winning Bid'], errors='coerce')
```

```python
df['sold_flag'] = np.where(df['Sold/Unsold'].str.lower() == 'sold', 1, 0)
```

```python
df
```

```python
df['Team'] = df['Team'].fillna('unsold')
#df['team'] = df['team'].replace({'nan':'Unsold'})
#df.loc[df['sold_flag'] == 0, 'team'] = df.loc[df['sold_flag'] == 0, 'team'].fillna('Unsold')
```

```python
df
```

```python
df['premium_paid'] = np.where(df['sold_flag'] == 1,
                              df['Winning Bid'].fillna(0) - df['Base Price'],
                              0)
```

```python
df
```

```python
df['bid_multiple'] = np.where(df['sold_flag'] == 1,
                              df['Winning Bid'] / df['Base Price'],
                              0)
```

```python
df
```

```python
df['bid_multiple'] = df['bid_multiple'].round(1)
```

```python
df
```

```python
df['buy_rating'] = np.where(
    df['sold_flag'] == 1,  # only for sold players
    pd.cut(
        df['Winning Bid'], 
        bins=[-1, 15000000, 50000000, np.inf],
        labels=['Budget Buy', 'Medium Buy', 'Premium Buy']
    ),
    np.nan  # unsold players stay null
)
```

```python
df
```

```python
column_names = df.columns.tolist()
print(column_names)
```

```python
sold=df[df['sold_flag'] == 1]
```

```python
sns.countplot(x='Sold/Unsold', data=df)

plt.title('Count of Sold vs Unsold Players')
plt.xlabel('Status')
plt.ylabel('Number of Players')
plt.show()
```

```python
sns.countplot(x='buy_rating',data=sold)

plt.title('Buy rating distribution')
plt.xlabel('Buy rating category')
plt.ylabel('Player Count')


plt.show
```

```python
bm = sold['bid_multiple']

plt.hist(bm, bins=20, label='Bid Multiple')

plt.title('Bid Multiple Distribution')
plt.xlabel('Bid Multiple')
plt.ylabel('Player Count')
plt.legend()
plt.show()
```

```python
sns.countplot(x='Capped/Uncapped', data=sold)

plt.title('Capped vs Uncapped — Sold Players')
plt.xlabel('Player Category')
plt.ylabel('Player Count')

plt.show()
```

```python
sns.countplot(x='Team', data=sold)

plt.title('Players Bought Per Team')
plt.xlabel('Team')
plt.ylabel('Player Count')
plt.xticks(rotation=45)
```

```python
team_spend = sold.groupby('Team')['Winning Bid'].sum()

plt.bar(team_spend.index, team_spend.values, label='Total Spend')

plt.title('Total Spend Per Team')
plt.xlabel('Team')
plt.ylabel('Total Winning Bid Amount')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
avg_spend = sold.groupby('Team')['Winning Bid'].mean().reset_index()

sns.barplot(x='Team', y='Winning Bid', data=avg_spend)

plt.title('Average Spend Per Player (Team-wise)')
plt.xlabel('Team')
plt.ylabel('Average Winning Bid')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='Team', hue='buy_rating', data=sold)

plt.title('Buy Rating Preference by Team')
plt.xlabel('Team')
plt.ylabel('Player Count')
plt.xticks(rotation=45)
plt.legend(title='Buy Rating')
plt.show()
```

```python
team_premium = sold.groupby('Team')['premium_paid'].sum()

plt.bar(team_premium.index, team_premium.values, label='Premium Paid')

plt.title('Total Premium Paid Per Team')
plt.xlabel('Team')
plt.ylabel('Premium Amount')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python

counts = df['sold_flag'].value_counts()

labels = ['Unsold', 'Sold']
sizes = counts.sort_index()

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Auction Outcome: Sold vs Unsold Players')
plt.show()
```

```python
rating_counts = sold['buy_rating'].value_counts()

plt.pie(
    rating_counts.values,
    labels=rating_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

# donut hole
centre_circle = plt.Circle((0,0), 0.6, color='white')
plt.gca().add_artist(centre_circle)

plt.title('Buy Rating Share (Sold Players)')
plt.show()
```

```python
plt.boxplot(sold['Winning Bid'])

plt.title('Winning Bid Price Spread')
plt.ylabel('Winning Bid Amount')
plt.show()
sns.boxplot(x='buy_rating', y='Winning Bid', data=sold)
```

```python
plt.boxplot(sold['premium_paid'])

plt.title('Premium Paid Spread')
plt.ylabel('Premium Paid Amount')

plt.show()
```

```python
top5 = sold.nlargest(5, 'Winning Bid')

plt.bar(top5['Player Name'], top5['Winning Bid'], label='Winning Bid')

plt.title('Top 5 Most Expensive Players')
plt.xlabel('Player Name')
plt.ylabel('Winning Bid')
plt.xticks(rotation=30)
plt.legend()

plt.show()
```
